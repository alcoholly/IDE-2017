```python
% matplotlib inline
import json
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division 
import pickle
import numpy as np
from networkx.readwrite import json_graph
```

```python
graph = pickle.load(open('music_Graph.json')) # this a NetworkX graph object

print 'Number of nodes:', len(graph.nodes())
print 'Number of edges:', graph.number_of_edges()
print ' These are the attributes attached to each node:', graph.node[ u'Brand New'].keys()
# node degree will be used to adjust the size of each node
# communities will be used to divide the nodes to differently colored groups 
# according to musical tags indicative of genres/styles
print 'For example:'
print 'Alicia Keys belongs to community:', graph.node[ u'Alicia Keys']['community']
print 'Alicia Keys has a degree of:', graph.node[ u'Alicia Keys']['node degree']
```

```python
# Now we convert the graph to json format to use it for D3 visualization:
#https://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.readwrite.json_graph.node_link_data.html
JS_graph= json_graph.node_link_data(graph)
JS_graph
# Note that the nodes are ok (the same format as in this link: https://bl.ocks.org/mbostock/4062045)
# but the Links are encoded in numbers -instead of using the artist names and i don't know if it will work as in the
# example in the link
```

```python
# Now we got it ! :D
links=[{'source': i[0], 'target': i[1]} for i in graph.edges()]
nodes= JS_graph['nodes']
graph_json={'nodes':nodes, 'links':links}
graph_json
```
